```python
# function converts salaries to floats, concerts YOY to float, and fixes duplicate city names
def clean_variables():
    df["Value"] = df[df.columns[5:6]].replace('[\$,]', '', regex=True)
    df["Value"] = df[df.columns[5:6]].replace('[\%,]', '', regex=True).astype(float)
    df["YoY"] = df[df.columns[6:]].replace('[\%,]', '', regex=True).astype(float)
    df["Metro"] = df["Metro"].replace("New-York-City", "New York City")
    df["Metro"] = df["Metro"].replace("San-Francisco", "San Francisco")
    df["Metro"] = df["Metro"].replace("Los-Angeles", "Los Angeles")
    df["Metro"] = df["Metro"].replace("Washington-DC", "Washington DC")
```

# Create datasets for job titles

```python
# function creates dataframes for job titles and exports as csv
# Adjusts the month and year to a numeric value
# One-hot encodes the metropolitan areas
# Creates target variable
def Job_Title_Prep(dataframe, value, filename):
    dataframe["Month"] = dataframe[dataframe.columns[2:3]].replace('[-]', '', regex=True).astype(float)
    Metro_dummies = pd.get_dummies(dataframe["Metro"])
    dataframe["Atlanta"] = Metro_dummies["Atlanta"]
    dataframe["Boston"] = Metro_dummies["Boston"]
    dataframe["Chicago"] = Metro_dummies["Chicago"]
    dataframe["Houston"] = Metro_dummies["Houston"]
    dataframe["Los Angeles"] = Metro_dummies["Los Angeles"]
    dataframe["National"] = Metro_dummies["National"]
    dataframe["New York City"] = Metro_dummies["New York City"]
    dataframe["Philadelphia"] = Metro_dummies["Philadelphia"]
    dataframe["San Francisco"] = Metro_dummies["San Francisco"]
    dataframe["Seattle"] = Metro_dummies["Seattle"]
    dataframe["Washington DC"] = Metro_dummies["Washington DC"]
    dataframe['Value_split'] = dataframe['Value']
    dataframe.loc[dataframe.Value_split < value, 'Value_split'] = 0
    dataframe.loc[dataframe.Value_split > value, 'Value_split'] = 1
    dataframe.to_csv(filename)
```

```python
# Run function on 12 job titles
Job_Title_Prep(df_RN,RN_mean,'df_RN.csv')
Job_Title_Prep(df_Data_Scientist,DataScientist_mean,'df_Data_Scientist.csv')
Job_Title_Prep(df_Accountant,Accountant_mean, 'df_Accountant.csv')
Job_Title_Prep(df_IT_Manager,ITMgr_mean,'df_IT_Manager.csv')
Job_Title_Prep(df_Sales_Rep,SalesRep_mean,'df_Sales_Rep.csv')
Job_Title_Prep(df_Professor,Professor_mean,'df_Professor.csv')
Job_Title_Prep(df_Teacher,Teadcher_mean,'df_Teacher.csv')
Job_Title_Prep(df_Software_Engineer,SoftwareEng_mean,'df_Software_Engineer.csv')
Job_Title_Prep(df_Project_Mgr,ProjectMgr_mean,'df_Project_Mgr.csv')
Job_Title_Prep(df_Financial_Analyst,FinAnalyst_mean,'df_Financial_Analyst.csv')
Job_Title_Prep(df_Data_Analyst,DataAnalyst_mean,'df_Data_Analyst.csv')
Job_Title_Prep(df_Graphic_Designer,GraphicDesigner_mean,'df_Graphic_Designer.csv')
```
